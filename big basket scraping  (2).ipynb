```python
def GetProductData(i):
    sleep(5)
    page_source=driver.page_source
    soup=BeautifulSoup(page_source,'lxml')
    bs4.BeautifulSoup
    title=soup.title
    
    #product name
    productname_h1Tag=soup.find("h1",{"class":"GrE04"})
    product_name.append(productname_h1Tag.text)
    
    
    #product brand name
    brandname_Tag=soup.find("a",{"context":"brandlabel"})
    brand_name.append(brandname_Tag.text)
    
    #product price 
    try:
        pricename_Tag=soup.find("tr",{"class":"_23Nyv"})
        price_list.append(pricename_Tag.text)
    except:
        price_list.append("NAN")
    
    #priduct final price
    try:
        pricefinal_Tag=soup.find("tr",{"class":"_157dw"})
        price_list_final.append(pricefinal_Tag.text)
    except:
        price_list_final.append("NAN")
    #PRIDUCT SAVE MONEY
    try:
        save_Tag=soup.find("tr",{"class":"_21awm"})
        save_money.append(save_Tag.text)
    except:
        save_money.append("NAN")
    
    #description
    
    decr_tag=soup.find("div",{"id":"about_0"})
    
    descrpt_list.append(decr_tag.text)
    
    
    #
    try:
        decrs_tag=soup.find("div",{"id":"about_1"})
        descrpt_list1.append(decrs_tag.text)
    except:
        descrpt_list1.append("NAN")
        
    
    try:
        how_tag=soup.find("div",{"id":"about_2"})
        how_to_use.append(how_tag.text)
    except:
        how_to_use.append("NAN")
        
    ImageScarpe(i,soup,productname_h1Tag.text)
        
        
    
        
        
    
    
```

```python
def ImageScarpe(imnumb,soup,productname):
    
    
    #get the image source
    
    try:
        div_tag = soup.find_all("div",{"class":"_2FbOx"})[1]
        productimage_tag=div_tag.find("img")
        
    except:
        div_tag = soup.find_all("div",{"class":"_2FbOx"})[0]
        productimage_tag=div_tag.find("img")
        
        
    image_url = productimage_tag["src"]
    
    #open the url image , set stream to True, this will return the stream content.
    resp=requests.get(image_url,stream=True)
    
    
    local_file=open("G:/Bigbasket Data/Imagescarpe//"+ productname.replace("/"," ") +".jpg","wb")
    
    
    #set decode_content value to true 
    resp.raw.decode_content= True
    
    
    #copy the response stream raw data to localimage file
    
    shutil.copyfileobj(resp.raw,local_file)
    
    del resp
```

```python
def Main(catgName):
    #print
    TotalProducts=ExtPoductlinks(catgName)
    
    loc=(r"G:\data\Bigbasket-Product.xlsx")
    wb=xlrd.open_workbook(loc)
    sheet=wb.sheet_by_index(0)
    sheet.cell_value(0,0)
    
    for i in range(1,TotalProducts+1):
        print(i)
        
        driver.get(sheet.cell_value(i,1))
        
        GetProductData(i)
        
    
    
    
```
